```python
data.drop(columns=['Male','Female'],inplace=True)
data.head()
```

```python
train_arr=data.values
train_arr.shape
```

```python
for i in range(len(train_arr[0])-1):
    
    mean = np.mean(train_arr[:,i])
    std = np.std(train_arr[:,i])
    train_arr[:, i] = (mean - train_arr[:,i]) / std
```

```python
train_arr
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
y = train_arr[:,-1]
```

```python
#y = data['pricing_category']
#train = data.drop(columns=['pricing_category'])
```

```python
train = train_arr[:,:-1]
```

```python
train.shape
```

```python
x_train,x_test,y_train,y_test = train_test_split(train,y,random_state=0,test_size=0.25)
```

```python
clf = LogisticRegression(random_state=0, solver='saga',multi_class='multinomial')
```

```python
clf.fit(x_train,y_train)
```

```python
ypred = clf.predict(x_test)
```

```python
clf.score(x_test,y_test)
```
